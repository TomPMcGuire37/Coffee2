```python
import numpy as np
import pandas as pd
```

I ask for inputs that then get placed into a dataframe before concatenating the new data to my existing csv. There should be more if else statements if this was going into a production but for my own personal use, I believe it is in an acceptable state.

```python
# Get the coffee info from that day

dummy_df = pd.DataFrame()

coffee_yn = input("Did you have coffee today? ")

coffee_yn = coffee_yn.lower()
if coffee_yn == 'y' or coffee_yn == 'yes':
    date_input = input('Enter the date: ')
    brand_input = input('Enter the brand of coffee used: ')
    roast_input = input('Enter the roast: ')
    blend_input = input('Enter the blend name: ')
    ice_input = input('Was the coffee made over ice? ')
    grind_input = input('What grind size was used? ')
    method_input = input('What method was used to brew? ')
    coffee_input = input('How much coffee in grams was used? ')
    water_input = input('How much water was used to brew? ')
    ice_amount = input('How much ice was used? ')
    bloom_input = input('How long was the coffee allowed to bloom? ')
    grinder_used = input('Which grinder was used?').strip() or 'Wilfa'
    total_time_minutes = input('How many minutes to fully brew coffee?')
    total_time_seconds = input('How many seconds to fully  brew coffee?')
    rating_input = input('What was the overall rating of the coffee? ')
    new_row = {'Date': date_input, 'Brand': brand_input, 'Roast':roast_input, 
               'Blend': blend_input, 'Iced': ice_input,
              'Grind_size': grind_input, 'Method': method_input, 'Amount_coffee': coffee_input,
               'Amount_water': water_input, 'Amount_ice': ice_amount, 'Bloom_time': bloom_input, 
               'Grinder_used': grinder_used, 'Total_time_m': total_time_minutes, 'Total_time_s':total_time_seconds, 'Rating': rating_input}

    #dummy_df = dummy_df.append(new_row, ignore_index = True)
    new_row = pd.DataFrame([new_row])
    dummy_df = pd.concat([dummy_df,new_row], ignore_index = True)

else:
    print('Have a good day')
```
